```python
# Can Gross profit of a movie be predicted with budget, genre, rating... etc. 
```

```python
# Read in data - downloaded from Kaggle https://www.kaggle.com/danielgrijalvas/movies
movies = pd.read_csv("/Users/camillahandley/Stat 426/Stat426FinalProject/Movies.csv")

# Only want US movies
movies = movies[movies['country'] == "USA"]
# Take out variables we don't need 
movies.drop(columns = ['runtime', 'score', 'votes', 'director', 'writer', 'country', 'star', 'company'], inplace = True)
```

```python
# Parse out the Month from 'released' variable
pattern = '\d+'
# assign False if there is not a full date and take out those observations 
movies = movies[movies['released'].apply(lambda x: True if len(re.findall(pattern, x)) == 3 else False)] 
movies['Month'] = movies['released'].apply(lambda x: int(re.findall(pattern, x)[1]))
# Remove extra columns 
movies.drop(columns = ['released'], inplace = True)
```

```python
# We only want movies that are rated (R, PG-13, PG, G)
rated = ['R', 'PG-13', 'PG', 'G']
# Take out the unrated observations
movies = movies[movies['rating'].isin(rated)]
```

```python
# Make dummy season variables
movies['winter'] = [1 if i == 12 or i == 1 or i == 2 else 0 for i in movies['Month']]
movies['spring'] = [1 if i > 3 and i < 6 else 0 for i in movies['Month']]
movies['fall'] = [1 if i > 8 and i < 12 else 0 for i in movies['Month']]
movies['summer'] = [1 if i > 5 and i < 9 else 0 for i in movies['Month']] 
movies.drop(columns = 'Month', inplace = True)
```

```python
# Explore the data with some EDA 
# Does profit increase with bigger budgets??
plt.scatter(movies['budget'], movies['gross'])
plt.xlabel("Movie Budget")
plt.ylabel('Gross Profit')
plt.title("Budget and Profit")
```

```python
# Does profit increase over the years?? 
plt.bar(movies.groupby('year').mean().index, movies.groupby('year').mean()['gross'])
plt.xlabel("Year Released")
plt.ylabel('Gross Profit')
plt.title("Average Profit from 1985 - 2016")
```

```python
# Do certain genres make more than others?
plt.bar(movies.groupby('genre').mean().index, movies.groupby('genre').mean()['gross'])
plt.xticks(rotation=40)
plt.xlabel("Genre")
plt.ylabel("Average Gross Profit")
```

```python
# Do certain ratings make more than others?
plt.bar(movies.groupby('rating').mean().index, movies.groupby('rating').mean()['gross'])
#plt.xticks(rotation=40)
plt.xlabel("Rating")
plt.ylabel("Average Gross Profit")
```

```python
# Read in the movie data from 2012-2012 to add a theater variable
table = pd.read_html("https://www.boxofficemojo.com/year/2016/?grossesOption=calendarGrosses")
m2016 = table[0]
m2016['year'] = 2016
table = pd.read_html("https://www.boxofficemojo.com/year/2015/?grossesOption=calendarGrosses")
m2015 = table[0]
m2015['year'] = 2015
table = pd.read_html("https://www.boxofficemojo.com/year/2014/?grossesOption=calendarGrosses")
m2014 = table[0]
m2014['year'] = 2014
table = pd.read_html("https://www.boxofficemojo.com/year/2013/?grossesOption=calendarGrosses")
m2013 = table[0]
m2013['year'] = 2013
table = pd.read_html("https://www.boxofficemojo.com/year/2012/?grossesOption=calendarGrosses")
m2012 = table[0]
m2012['year'] = 2012


theaters = pd.concat([m2016, m2015, m2014, m2013, m2012], ignore_index = True)
```

```python
# Remove empty columns and ones we won't be using
theaters.drop(columns = ['Genre', 'Budget', "Running Time", 'Estimated', 'Release Date',
                      'Gross', 'Rank', 'Total Gross'], inplace = True)
theaters.shape
```

```python
# Take out all the extra years in the movie dataset (we only want 2012-2016)
movies = movies.loc[movies['year'] > 2011]

# Rename the Release column that that is matches with the movies dataset 
theaters.rename(columns={"Release": "name"}, inplace = True)
theaters.head()
```

```python
# This decreases the size of our dataset by a lot but we only want movies that are in both datasets 
movies = pd.merge(movies, theaters, on = ["name",'year'], how = 'inner')
movies['Theaters'] = movies['Theaters'].apply(lambda x: int(x))
```

```python
movies.head()
```

```python
# Split into test and training set 

# Use data from years 2012-2015 to predict gross from 2016
Testdata = movies.loc[movies['year'] == 2016]
Traindata = movies.loc[movies['year'] < 2016]

X_train = Traindata[['budget', 'genre', 'rating', 'year', 'winter', 'spring', 'fall', 'summer', 'Theaters']]
y_train = Traindata.loc[:,'gross']

X_test = Testdata[['budget', 'genre', 'rating', 'year', 'winter', 'spring', 'fall', 'summer', 'Theaters']]
y_test = Testdata.loc[:,'gross']
```

```python

class ColumnExtractor(TransformerMixin):
    
    def __init__(self, cols):
        self.cols = cols
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        Xcols = X[self.cols]
        return Xcols

#  Numerical features
numeric_columns = X_train.select_dtypes(exclude='object').columns

#  Categorical features
categorical_columns = X_train.select_dtypes('object').columns

#  Define the Pipeline
pipeline = Pipeline([
    ('features', FeatureUnion([
        ('numeric', Pipeline([
            ('extract', ColumnExtractor(numeric_columns)),
            ('impute', SimpleImputer())
        ])), # Make a dummy variable for rating and genre and company
        ('categorical', Pipeline([
            ('extract', ColumnExtractor(categorical_columns)),
            ('impute', SimpleImputer(strategy='constant', fill_value='None')),
            ('encode', OneHotEncoder(handle_unknown='ignore', sparse=False))
        ])),
    ])),
    ('scale', StandardScaler())  #  scale all features
])


lasso_pipeline = Pipeline([('pipeline', pipeline),
                         ('lasso', Lasso(alpha = 0.5))])
knn_pipeline = Pipeline([('pipeline', pipeline),
                         ('knn', KNeighborsRegressor(n_neighbors=3))])
lm_pipeline = Pipeline([('pipeline', pipeline),
                         ('lm', LinearRegression())])
```

```python
# Linear Model 
lm_pipeline.fit(X_train, y_train)
yhat_train = lm_pipeline.predict(X_train)
rmse = np.sqrt(mean_squared_error(y_train, yhat_train))
score = lm_pipeline.score(X_train, y_train)

#Showing the RMSE for this model
print("RMSE (Training) : {:.3f}".format(rmse))

# Showing the accuracy
print("Accuracy on training set: {:.3f}".format(score))
```

```python
# Lasso Regression 
lasso_pipeline.fit(X_train, y_train)
yhat_train = lasso_pipeline.predict(X_train)
rmse = np.sqrt(mean_squared_error(y_train, yhat_train))
score = lasso_pipeline.score(X_train, y_train)

#Showing the RMSE for this model
print("RMSE (Training) : {:.3f}".format(rmse))

# Showing the accuracy
print("Accuracy on training set: {:.3f}".format(score))
```

```python
# K Nearest Neighbors
knn_pipeline.fit(X_train, y_train)
yhat_train = knn_pipeline.predict(X_train)
rmse = np.sqrt(mean_squared_error(y_train, yhat_train))
score = knn_pipeline.score(X_train, y_train)

#Showing the RMSE for this model
print("RMSE (Training) : {:.3f}".format(rmse))

# Showing the accuracy
print("Accuracy on training set: {:.3f}".format(score))
```

```python
# Solely based on these two metrics, the KNN model performs the best 
# It has the highest accuracy and the lowest RMSE 
```

```python
# See what # of neighbors would be best 
mse_cv = []
mse_cv_mean = []

for p in np.arange(1,20):
    knn_pipe = Pipeline([('pipeline', pipeline),
                         ('knn', KNeighborsRegressor(n_neighbors=p))])
    knn_pipe.fit(X_train, y_train)
    val = mean_squared_error(y_test, knn_pipe.predict(X_test))
    mse_cv.append(val)
```

```python
plt.plot(np.arange(1,20), mse_cv)
# Based off this graph it looks like 5 would be the best 
```

```python
# Using the optimal # of neighbors (for RMSE)
knn_pipe = Pipeline([('pipeline', pipeline),
                         ('knn', KNeighborsRegressor(n_neighbors=5))])

# Use to predict 2016 gross profits 
knn_pipe.fit(X_train, y_train)
yhat_test = knn_pipe.predict(X_test)
# Error is true value - prediction
error = y_test - yhat_test
Testdata['error'] = error
```

```python
plt.scatter(x = np.arange(1, len(error)+1), y = error)
plt.xlabel("Index")
plt.ylabel("Prediction Error")
plt.title("Prediction Error for 2016 Gross Profit")
```
