```python
# Final Project # 
# Stat 426 - Dr. Tass # 

import pandas as pd 
import numpy as np 
import re
```

```python
# Read in the movie data from 2016-2019 and Add a year variable

table = pd.read_html("https://www.boxofficemojo.com/year/2019/?grossesOption=calendarGrosses")
movies = table[0]
movies['year'] = 2019
table = pd.read_html("https://www.boxofficemojo.com/year/2018/?grossesOption=calendarGrosses")
m2018 = table[0]
m2018['year'] = 2018
table = pd.read_html("https://www.boxofficemojo.com/year/2017/?grossesOption=calendarGrosses")
m2017 = table[0]
m2017['year'] = 2017
table = pd.read_html("https://www.boxofficemojo.com/year/2016/?grossesOption=calendarGrosses")
m2016 = table[0]
m2016['year'] = 2016
table = pd.read_html("https://www.boxofficemojo.com/year/2015/?grossesOption=calendarGrosses")
m2015 = table[0]
m2015['year'] = 2015
movies = pd.concat([movies, m2018, m2017, m2016, m2015], ignore_index = True)
# Remove empty columns 
movies.drop(columns = ['Genre', 'Budget', "Running Time", 'Estimated'], inplace = True)
movies.shape
```

```python
# Read in data - downloaded from Kaggle https://www.kaggle.com/danielgrijalvas/movies
movies = pd.read_csv("/Users/camillahandley/Stat 426/Stat426FinalProject/Movies.csv")

# Only want US movies
movies = movies[movies['country'] == "USA"]
# Take out variables we don't need 
movies.drop(columns = ['runtime', 'score', 'votes', 'director', 'writer', 'country', 'star'], inplace = True)
```

```python
# Parse out the Month and Day from 'released' variable
pattern = '\d+'
# assign False if there is not a full date and take out those observations 
movies = movies[movies['released'].apply(lambda x: True if len(re.findall(pattern, x)) == 3 else False)] 
movies['Day'] = movies['released'].apply(lambda x: int(re.findall(pattern, x)[2]))
movies['Month'] = movies['released'].apply(lambda x: int(re.findall(pattern, x)[1]))
# Remove extra columns 
movies.drop(columns = ['released'], inplace = True)
```
